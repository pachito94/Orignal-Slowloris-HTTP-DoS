###3. Numpy manejo numerico

```python
import numpy as np
habitantes =np.array([130000,20000,15000,12000,18000,8000,10000,7000])
consumos=np.array([62000,13000,11000,9000,12000,5000,6000,4000])
print("media habitantes:", np.mean(habitantes))
print("media de consumo:", np.mean(consumos))
```

###ðŸ–Š Ejemplo: Calcular la correlacion de habitantes y consumo

```python
correlacion=np.corrcoef(habitantes,consumos)[0,1]
print("correlacion habitantes consumo:", round(correlacion,2))
```

ðŸ§ Interpretacion: un valor cercano a 1 significa que a mas habitantes, mas consumo valor cercano a 0 indica que no hay relacion
